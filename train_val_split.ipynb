```python
def split_train_validation(data, validation_ratio=0.2):
    """
    Splits the data into training and validation sets according to the specified validation_ratio.
    The last X% of rows are assigned to the validation set.
    
    Parameters:
        data (pd.DataFrame): the full dataset
        validation_ratio (float): the proportion of data to use for validation (default is 0.2)

    Returns:
        train_data (pd.DataFrame), val_data (pd.DataFrame)
    """
    split_index = int((1 - validation_ratio) * len(data))
    train_data = data.iloc[:split_index]
    val_data = data.iloc[split_index:]
    return train_data, val_data
```

```python
train, val = split_train_validation(train_prev, validation_ratio=0.2)
```
