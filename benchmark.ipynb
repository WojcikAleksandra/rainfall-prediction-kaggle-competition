```python
import pandas as pd
```

# Upload data

```python
X_train = pd.read_csv('data/to_model/X_train.csv')
X_valid = pd.read_csv('data/to_model/X_val.csv')
y_train = pd.read_csv('data/to_model/Y_train.csv')
y_valid = pd.read_csv('data/to_model/Y_val.csv')
X_train.head()
```

**Modeling**
We will use random forest on this dataset to see how it performs.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

rf = RandomForestClassifier(n_estimators=300, random_state=42)
rf.fit(X_train, y_train.values.reshape(-1))
Y_pred = rf.predict_proba(X_valid)[:,1]
roc_auc_score(y_valid, Y_pred)
```
