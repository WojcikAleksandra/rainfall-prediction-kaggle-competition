```python
import pandas as pd
```

```python
X_train = pd.read_csv('data/X_train.csv')
Y_train = pd.read_csv('data/Y_train.csv')
X_val = pd.read_csv('data/X_valid.csv')
Y_val = pd.read_csv('data/Y_valid.csv')
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01, random_state=42)
predictions = iso_forest.fit_predict(X_train)
mask = predictions != -1 
X_train_cleaned = X_train[mask].reset_index(drop=True)
Y_train_cleaned = Y_train[mask].reset_index(drop=True)

print(f"Length of the original data: {len(X_train)}")
print(f"After Isolation Forest: {len(X_train_cleaned)}")
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01, random_state=42)
predictions = iso_forest.fit_predict(X_val)
mask = predictions != -1 
X_val_cleaned = X_val[mask].reset_index(drop=True)
Y_val_cleaned = Y_val[mask].reset_index(drop=True)

print(f"Length of the original data: {len(X_val)}")
print(f"After Isolation Forest: {len(X_val_cleaned)}")
```

### LightGBM

```python
import lightgbm as lgb

params = {
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'verbose': -1
}

train_data1 = lgb.Dataset(X_train, label=Y_train)
valid_data1 = lgb.Dataset(X_val, label=Y_val, reference=train_data1)

model1 = lgb.train(params, train_data1, valid_sets=[valid_data1], early_stopping_rounds=50, verbose_eval=False)

train_data2 = lgb.Dataset(X_train_cleaned, label=Y_train_cleaned)
valid_data2 = lgb.Dataset(X_val_cleaned, label=Y_val_cleaned, reference=train_data2)

model2 = lgb.train(params, train_data2, valid_sets=[valid_data2], early_stopping_rounds=50, verbose_eval=False)

y1_pred = model1.predict(X_val)
y2_pred = model2.predict(X_val_cleaned)

y1_pred_class = (y1_pred > 0.5).astype(int)
y2_pred_class = (y2_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import f1_score, roc_auc_score, precision_score, recall_score

metrics_1 = {
    'F1-score': f1_score(Y_val, y1_pred_class),
    'AUC': roc_auc_score(Y_val, y1_pred),
    'Precision': precision_score(Y_val, y1_pred_class),
    'Recall': recall_score(Y_val, y1_pred_class),
}

metrics_2 = {
    'F1-score': f1_score(Y_val_cleaned, y2_pred_class),
    'AUC': roc_auc_score(Y_val_cleaned, y2_pred),
    'Precision': precision_score(Y_val_cleaned, y2_pred_class),
    'Recall': recall_score(Y_val_cleaned, y2_pred_class),
}

print("Results for dataset with outliers:", metrics_1)
print("Results for dataset without outliers:", metrics_2)
```
